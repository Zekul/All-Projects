```python
import urllib.request
from bs4 import BeautifulSoup
import spacy
import neuralcoref
nlp = spacy.load('en_core_web_sm')
neuralcoref.add_to_pipe(nlp)
'''Use the following if you want to directly read a url text'''
html = urllib.request.urlopen('https://www.nzherald.co.nz/travel/hiker-finds-438-carat-diamond-in-us-park-is-hers-to-keep/NV6BVUTFA4MTORIUDC23GNKO5M')
soup = BeautifulSoup(html, 'html.parser')
text = ''.join([t for t in soup.find_all(text=True) if t.parent.name == 'p' and len(t) >= 25])
''' Use the following if you want to read the text from a file'''
# with open('data/sampleText2.txt', 'r',encoding='ascii', errors='ignore') as f:
#      text = f.read()


doc = nlp(text)
resolved_text = doc._.coref_resolved
print(text)
print('************************************************************************************')
#print(resolved_text)
#print(doc._.coref_clusters)

for ref in doc._.coref_clusters:
    print(ref)
print('************************************************************************************')
sentences = [sent.string.strip() for sent in nlp(resolved_text).sents]
output = [sent for sent in sentences if 'diamond' in
          (' '.join([token.lemma_.lower() for token in nlp(sent)]))]
print('Fact count:', len(output))
for fact in range(len(output)):
    print(str(fact+1)+'.', output[fact])
```
